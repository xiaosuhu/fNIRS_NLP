```python
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_td = scipy.io.loadmat("F:\\MatlabBackUp\\PROJECT_HH_ENGLISH\\Pilot_Aanalysis_LLM_data_prep\\brain_score_result_v2_TD.mat")
result_wd = scipy.io.loadmat("F:\\MatlabBackUp\\PROJECT_HH_ENGLISH\\Pilot_Aanalysis_LLM_data_prep\\brain_score_result_v2_WD.mat")
```

```python
avg_result_td = np.mean(result_td["brain_score"], axis=0)
avg_result_wd = np.mean(result_wd["brain_score"], axis=0)

norm_avg_result_td = avg_result_td #- avg_result_td[:,0][:,np.newaxis]
norm_avg_result_wd = avg_result_wd #- avg_result_wd[:,0][:,np.newaxis]
```

```python
# Step 3: Plot
fig, axes = plt.subplots(4, 5, figsize=(15, 10))  # Create a 4x5 grid of subplots
fig.suptitle("Channel-wise Time Series")

time = np.arange(16)  # Assuming time points are 0 to 15

for i, ax in enumerate(axes.flat):
    if i < 20:
        ax.plot(time, norm_avg_result_td[i, :], label=f"CH {i+1}")
        ax.plot(time, norm_avg_result_wd[i, :], label=f"CH {i+1}")

        ax.set_title(f"CH {i+1}")
        ax.set_xlabel("Time")
        ax.set_ylabel("Value")
        ax.grid(True)
        ax.legend(["TD","WD"])  # Add legend to each subplot

    else:
        ax.axis("off")  # Hide unused subplots

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit title
plt.show()
```
