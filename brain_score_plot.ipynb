```python
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_mat = scipy.io.loadmat("F:\\MatlabBackUp\\PROJECT_HH_ENGLISH\\Pilot_Aanalysis_LLM_data_prep\\brain_score_result_v2_WD.mat")
```

```python
result_mat["brain_score"].shape
```

```python
avg_result = np.mean(result_mat["brain_score"], axis=0)
```

```python
avg_result.shape
```

```python
norm_avg_result = avg_result - avg_result[:,0][:,np.newaxis]
```

```python
avg_result[:,0].shape
```

```python
norm_avg_result[17,:]
```

```python
# Step 3: Plot
fig, axes = plt.subplots(4, 5, figsize=(15, 10))  # Create a 4x5 grid of subplots
fig.suptitle("Channel-wise Time Series")

time = np.arange(16)  # Assuming time points are 0 to 15

for i, ax in enumerate(axes.flat):
    if i < 20:
        ax.plot(time, norm_avg_result[i, :], label=f"CH {i+1}")
        ax.set_title(f"CH {i+1}")
        ax.set_xlabel("Time")
        ax.set_ylabel("Value")
        ax.grid(True)
    else:
        ax.axis("off")  # Hide unused subplots

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit title
plt.show()
```
